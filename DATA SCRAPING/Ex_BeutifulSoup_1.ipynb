```python
from bs4 import BeautifulSoup

with open('index.html', 'r') as f:
    doc = BeautifulSoup(f, 'html.parser')
```

```python
print (doc.prettify())
#Here we can see the chematic 'Tree n Branch' in HTML. <html> child is <head>, <title> parent is <head> 
```

# Tag Name

```python
##TAG NAME
tag = doc.title #To know about the tag variables, just go to the tag name
print(tag)
print(tag.string) #to know value pr text inside the specific tag
```

```python
tag.string = 'hello' #We can change the value or text inside the tag. "From Your Title here" to "hello"
print(doc.prettify()) #N it also changed in the doc
```

# URL and Get the HTML of it

```python
import requests
```

```python
#Now we can grab the HTML of the website. But, there will also some website that we are not able to grab the HTML (e.g. Amazon)
url = 'https://www.newegg.ca/gigabyte-geforce-rtx-3080-ti-gv-n308tgaming-oc-12gd/p/N82E16814932436?Description=3080&cm_re=3080-_-14-932-436-_-Product'
result = requests.get(url)
print(result.text)
```

```python
#Now we can BeautifulSoup it
doc1 = BeautifulSoup(result.text, 'html.parser')
print(doc1.prettify())
```

# We want to locate and grab the price

```python
prices = doc1.find_all(text='$')
print(prices)
```

```python
parent = prices[1].parent
print(parent)
#To grab the value of price, we need to access the <strong>
```

```python
strong = parent.find('strong')
print(strong)
print(strong.text)
```
