```python
import PyPDF2 as pypdf
```

## Single Row

```python
pdf_path = 'Data/straight_table.pdf'
```

```python
with open(pdf_path,'rb') as pdffile:
    reader = pypdf.PdfFileReader(pdffile)
    for i in range(len(reader.pages)):
        page = reader.pages[i]
        text = page.extract_text()
        print('\n')
        print('PAGE_{}'.format(i))
        print('\n')
        print(text)
```
