```python
for i in range(len(dfs)):
    locals()['data_'+str(i)] = dfs[i].to_dict()
    locals()['df_'+str(i)] = pd.DataFrame.from_dict(locals()['data_'+str(i)])
```

```python
col_df=[]
for i in range(len(dfs)):
    
    #Auto Adjusment in the Columns
    try:
        while locals()['df_'+str(i)].columns[0] != 'Fascia d\'età':
            locals()['df_'+str(i)].columns = locals()['df_'+str(i)].iloc[0]
            locals()['df_'+str(i)] = locals()['df_'+str(i)][1:]
        
    except:
        print('df_'+str(i)+' is terminated')
        del locals()['df_'+str(i)]
     
    #Auto Rid the un-useful table (no data)
    try:
        if locals()['df_'+str(i)].columns[0] != 'Fascia d\'età':
            print('df_'+str(i)+' is terminated')
            del locals()['df_'+str(i)].columns[0]
    
    except:
        print('df_'+str(i)+' is ready')
    
    #Reset-Indexing in the DataFrame
    try:
        locals()['df_'+str(i)].reset_index(drop=True)
        col_df.append(locals()['df_'+str(i)])
    
    except:
        pass
```
