```python
from bs4 import BeautifulSoup
import requests

url = 'https://coinmarketcap.com/'
result = requests.get(url).text
doc = BeautifulSoup(result, 'html.parser')

#tbody is the parent for the table (which consisy of the name of all coins) that we want to inspect
tbody = doc.tbody
print(tbody)
```

```python
#Make a list of all tags inside the tbody. Trs = table rows
trs = tbody.contents

#'next_sibling' means that we search for the table row after trs[0] or row 1. It is proven by the coin name "ETH"
#'previous_sibling' can also be the case. The name willbe 'BTC'
#with 'next_siblings', we can get every rows after the [0]
print(trs[0].next_sibling)
```

```python
#We can also know about the parent of our trs, which is tbody by saying
print(trs[0].parent.name)

#And we can also know the descendant of our tag of trs by saying. In the format of generator because it has a lot of tags
#This is alternative for .contents
print(trs[0].descendants)

#With '.childern' will give us only the name of the tags
print(trs[0].children)
```

# Grab the price of the coins

```python
prices = {}

for tr in trs:
    for td in tr.contents[2:4]: #Now, we only slicing the data for the name and the price of the coin
        print(td)
        print()
```

```python
for tr in trs:
    name, price = tr.contents[2:4] #Now, we only calling the data for the name of the coin
    print(name)
    print()
```

```python
for tr in trs[:10]: #Only for 10
    name, price = tr.contents[2:4] #Now, we only calling the data for the name of the coin
    print(name.p.string) #by calling the <p> tag only, we can get the coin name
    print()
```

```python
for tr in trs[:10]: #Only for 10
    name, price = tr.contents[2:4] #Now, we only calling the data for the name of the coin
    fixed_name = name.p.string #by calling the <p> tag only, we can get the coin name
    print(price.a.string)
```
