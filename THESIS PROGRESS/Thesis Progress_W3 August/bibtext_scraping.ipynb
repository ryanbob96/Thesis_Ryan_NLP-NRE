```python
from pybtex.database.input import bibtex
import bibtexparser
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open("References_P3.bib",encoding='utf-8') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
```

```python
catalog_publisher = []
list_publisher = []
list_doi = []
list_title = []
none = 0
for i in bib_database.entries:
    try:
        list_publisher.append(i['publisher'])
        if i['publisher'] not in catalog_publisher:
            catalog_publisher.append(i['publisher'])
        else:
            pass
    except:
        none +=1
        
for i in bib_database.entries:
    try:
        list_title.append(i['title'])
        list_doi.append(i['doi'])
    except:
        pass
```

```python
dict_publisher = {}
for i in catalog_publisher:
    dict_publisher[i] = list_publisher.count(i)
```
