# Scraping: Geosphere from GeoscienceWorld

```python
path = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(path)
link_doi_geosphere = []
#Open up the website
driver.get('https://pubs.geoscienceworld.org/geosphere/list-of-years')
d = driver.find_element(By.CLASS_NAME, "widget.widget-ContentBrowseAllYearsManifest.widget-instance-IssueBrowseAllYears").text.split('\n')
for i in d:
    try:
        driver.implicitly_wait(3)
        driver.find_element(By.LINK_TEXT, i).click()
        driver.implicitly_wait(3)
    except:
        print('failed in ' + i)
    driver.implicitly_wait(3)
    c = driver.find_element(By.CLASS_NAME, "widget.widget-ContentBrowseByYearManifest.widget-instance-IssueBrowseByYear").text.split('\n')
    driver.implicitly_wait(3)
    for j in c:
        try:
            driver.implicitly_wait(3)
            driver.find_element(By.LINK_TEXT, j).click()
            driver.implicitly_wait(3)
        except:
            print('failed in ' + j)
        driver.implicitly_wait(3)
        f = driver.find_elements(By.CLASS_NAME, 'al-cite-description')
        driver.implicitly_wait(3)
        for t in f:
            try:
                link_doi_geosphere.append(regex.findall(r'doi:([\w\W]+)$',t.text)[0])
            except:
                print('failed in ' + t.text)
        driver.implicitly_wait(3)
        driver.back()
        driver.implicitly_wait(3)
    driver.implicitly_wait(3)
    driver.back()
    driver.implicitly_wait(3)
```

# Scraping: Springer

```python
path = 'C:\Program Files (x86)\chromedriver.exe'
driver = webdriver.Chrome(path)
#Open up the website
driver.get('https://www.springer.com/gp/journal-impact/earth-sciences-geography')

#Cookies
driver.find_element(By.CLASS_NAME, 'cc-button.cc-button--link.cc-button--contrast.cc-banner__button.cc-banner__button-preferences').click()
driver.find_element(By.CLASS_NAME, 'cc-button.cc-button--small').click()


list_successed_journal_extra = []
list_failed_journal_extra = []
List_failed_Volume_extra = []
List_failed_paper_extra = []
List_success_paper_extra = []
# a = driver.find_elements(By.CSS_SELECTOR, 'h3')
m = driver.find_elements(By.CLASS_NAME, 'product-information ')
a = []
for n in m:
    g = n.text.split('\n')
    if len(g) == 1:
        a.append(g[0])
for c in a[25:]:
    try:
        try:
            driver.implicitly_wait(5)
            driver.find_element(By.PARTIAL_LINK_TEXT, c).click()
            driver.find_element(By.CLASS_NAME, 'u-button').click()
            driver.implicitly_wait(5)
            d = driver.find_elements(By.CLASS_NAME, 'u-interface-link.u-text-sans-serif.u-text-sm')
            list_successed_journal_extra.append(c.text)
        except:
            pass

        try:
            b = [i.text for i in d]
        except:
            pass

        for k in b:
            try:
                driver.implicitly_wait(5)
                driver.find_element(By.PARTIAL_LINK_TEXT, k).click()
                driver.implicitly_wait(5)
                m = driver.find_elements(By.CSS_SELECTOR, 'h3.c-card__title [href]')
            except:
                List_failed_Volume_extra.append(k)

            for n in m:
                try:
                    driver.implicitly_wait(5)
                    List_success_paper_extra.append(n.get_dom_attribute('href'))
                except:
                    List_failed_paper_extra.append(n.get_dom_attribute('href'))
            driver.implicitly_wait(5)
            driver.back()
        driver.implicitly_wait(5)
        driver.get('https://www.springer.com/gp/journal-impact/earth-sciences-geography')
    except:
        list_failed_journal_extra.append(c)
        driver.implicitly_wait(5)
        driver.get('https://www.springer.com/gp/journal-impact/earth-sciences-geography')
```

```python
with open('list_doi_springer_extra.txt', 'w', encoding='utf-8') as f:
    for i in List_success_paper_extra:
        f.write(i)
        f.write('\n')
    f.close()  
```

```python
success_springer = []
with open('extracted_springert_Jimpact_text.txt', 'w', encoding='utf-8') as f:
    path = 'C:\Program Files (x86)\chromedriver.exe'
    driver = webdriver.Chrome(path)
    #Open up the website
    for i in List_success_paper_extra:
        time.sleep(2)
        driver.implicitly_wait(4)
        driver.get(i)
        time.sleep(2)
        driver.implicitly_wait(4)
        try:
            a = driver.find_element(By.CLASS_NAME, 'main-content').text
            f.write(a)
            f.write('\n')
            f.write('----END----')
            f.write('\n')
            f.write('----NEW----')
            f.write('\n')
            success_springer.append(i)
        except:
            print('no')
    f.close()
```
