## Data Pre-Processing (Stemming, Stopwords)

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

## OHE Representation

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
onehot_repr
```

```python
onehot_repr[2]
```

```python
corpus[2]
```

## Add Padding

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen=sent_length)
print(embedded_docs)
```

## OHE to Vector + Model

```python
embedding_vector_features=40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model.add(LSTM(100)) #LSTM with 100 neurons
model.add(Dense(1, activation = 'sigmoid')) #Our Output is 1 because its binary, and also activation of sigmoid
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model.summary())
```

## Adding Dropout

```python
from tensorflow.keras.layers import Dropout
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Dropout(0.3)) #After Embedding Layer, we can use dropout 30% of the neuron
model.add(LSTM(100))
model.add(Dropout(0.3)) #After LSTM, we can use dropout 30% of the neuron
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

## Model Training after Dropout (to avoid over-fitting)

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs=10, batch_size=6)
```
