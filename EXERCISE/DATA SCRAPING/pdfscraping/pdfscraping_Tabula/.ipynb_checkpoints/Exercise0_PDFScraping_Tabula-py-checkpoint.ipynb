```python
import tabula
import pandas as pd
```

```python
pdf_path = 'Data/ast_sci_data_tables_sample.pdf'
```

# Single Table

```python
dfs = tabula.read_pdf(pdf_path, pages='1', stream=True)
print(dfs)
print(len(dfs))
print(dfs[0])
```

```python
dfs[0].to_csv('Data/table_1.csv')
```

```python
#We can also make these in 1 line
#But it's only good for 1 table
tabula.convert_into(pdf_path,'Data/table_1_2.csv',output_format='csv',pages=1 )
```

# Multiple Table

```python
dfs = tabula.read_pdf(pdf_path, pages='2', stream=True)
print(dfs)
```

```python
#Multiple Table into csv
for i in range(len(dfs)):
    dfs[i].to_csv(f'table_2_part{i}.csv')
```

```python
#Multiple Table into library, into pandas dataframe
for i in range(len(dfs)):
    locals()['data_'+str(i)] = dfs[i].to_dict()
    locals()['df_'+str(i)] = pd.DataFrame.from_dict(locals()['data_'+str(i)])
```
